```python
import pymorphy2
mor = pymorphy2.MorphAnalyzer()
```

```python
!ls
import pandas as pd
data = pd.read_excel('cloze_task_average_accuracy.xlsx')
```

```python
def get_tag(x): 
    return mor.parse(x)[0].tag.POS

data['t_form'] = data['word.id'].apply(get_tag)
data['t_lemms'] = data['Lemma'].apply(get_tag)
data.head()
```

```python
writer = pd.ExcelWriter('output.xlsx')
data.to_excel(writer,'Sheet1')
writer.save()
```
